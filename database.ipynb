```python
import csv
reslut = []
with open('/product_list.csv', 'r', encoding='euc-kr') as f:
    reader = csv.reader(f)
    for line in reader:
        reslut.append(line)
        print(line)
```

```python
#csv 파일 수정 1
with open('/product_list.csv', 'w', encoding= 'euc-kr', newline= '') as f:
    writer = csv.writer(f)
    writer.writerows(reslut)               # 이전 레코드
    writer.writerow(['10', ' 운동화', '8000']) # 신규 레코드
```

```python
#csv 파일 수정 2
reslut.append(['11', '슬리퍼', '3000'])

with open('/product_list.csv', 'w', encoding= 'euc-kr', newline= '') as f:
    writer = csv.writer(f)
    writer.writerows(reslut)
```

# 데이터 삽입

```python
conn = sqlite3.connect('/test.db')
```

```python
c = conn.cursor()
```

```python
query = '''CREATE TABLE test(ID INTEGER PRIMARY KEY, PRODUCT_NAME TEXT, PRICE INTEGER)'''
c.execute(query)
```

```python
c.execute('''SELECT * FROM sqlite_master WHERE type= "table"''')
print(c.fetchall())
```

```python
# 데이터 삽입 키 순서를 알고 있을때
c.execute("INSERT INTO test VALUES(1, '모자', 15000)")
conn.commit()
```

```python
# 순서를 모르는 경우
c.execute("INSERT INTO test(PRODUCT_NAME, PRICE, ID) VALUES(?,?,?)", ('티셔츠', 20000, 3))
conn.commit()
```

```python
# 여러 데이터 삽입
product_list = [[1, '모자', 12000], [2, '코트', 200000], [3, '티셔츠', 15000], [4, '구두', 30000]]
c.executemany("INSERT INTO test(ID, PRODUCT_NAME, PRICE) VALUES(?,?,?)", product_list)
conn.commit()
```

# 데이터 수정

```python
# 튜플 형태로 수정
c.execute("UPDATE test SET PRODUCT_NAME = ? WHERE ID =?", ('슬랙스', 4))
conn.commit()
```

```python
# 딕셔너리 형태로 수정
c.execute("UPDATE test SET PRICE = :price WHERE ID = :id", {"price":55000, "id":1})
conn.commit()
```

```python
# %s 표시자 사용
c.execute("UPDATE test SET PRODUCT_NAME = '%s' WHERE ID ='%s'" %('트랜티코트', 2))
conn.commit()
```

# 데이터 삭제

```python
# 테이블 내용 삭제
c.execute('DELETE From test')
conn.commit()
```

```python
# 튜플 형태로 삭제 삭제
c.execute('DELETE From test WHERE ID =?', (3, ))
conn.commit()
```

```python
# 딕셔너리 형태로 삭제
c.execute("DELETE from test WHERE PRODUCT_NAME = :product_name", {"product_name":'슬랙스'})
conn.commit()
```

# 데이터 백업하기

```python
for line in conn.iterdump():
  print(line)
```

```python
# 백업파일 생성
with conn:
  with open('backup sql', 'w') as f:
    for line in conn.iterdump():
      f.write('%s\n' % line)
    print('completed')
```

```python
# 테이블 삭제
c.execute('drop table test')
conn.commit()
```

```python
# 백업 sql 파일 로딩
with open('backup sql', 'r') as sql_file:
  sql_script = sql_file.read()
```

```python
sql_script
```

```python
#sql 스크랩트 실행
c.executescript(sql_script)
conn.commit()
```
